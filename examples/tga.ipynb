```python
selected_columns = [
    2,
    1,
    5,
    3,
    4,
]  # column positions of time, sample temperature, reference temperature, heat flow and weight, repsectively.
selected_df = df.iloc[:, selected_columns]
# New column names (tuples for MultiIndex)
new_header_names = [
    ("Time", "[s]"),
    ("Ts", "[°C]"),
    ("Tr", "[°C]"),
    ("HF", "[mW]"),
    ("Weight", "[mg]"),
]

# Assign the new column names to the DataFrame
selected_df.columns = pd.MultiIndex.from_tuples(new_header_names)
selected_df.head(20).style
```

```python
selected_df = selected_df.dropna()
selected_df = selected_df.round(3)
selected_df = selected_df.astype(float)
selected_df[("Time", "[min]")] = selected_df[("Time", "[s]")] / 60
selected_df[("Ts", "[K]")] = selected_df[("Ts", "[°C]")] + 273
selected_df[("Tr", "[K]")] = selected_df[("Tr", "[°C]")] + 273
selected_df[("HF", "[mW/mg]")] = (
    selected_df[("HF", "[mW]")] / selected_df[("Weight", "[mg]")]
)
selected_df.head(30).style
```

```python
target_column = selected_df[("Tr", "[°C]")]
target_index = target_column[target_column == 30].index[0]
selected_df = selected_df.loc[target_index:]
selected_df[("Weight", "[%]")] = (
    selected_df[("Weight", "[mg]")] / selected_df[("Weight", "[mg]")].iloc[0] * 100
)
selected_df.head(30).style
```
